### LRP with ESN trained on synthetic samples

Start with fading memory experiments on ESN, feeding samples column-wise into the model. Apply various leak rates:
- 0.005
- 0.05
- 0.2

Then keep leak rate $\alpha=0.005$ fixed and compare for ESN:
- feed samples column-wise
- feed samples row-wise
- feed samples as equal-sized pieces

### Load and prepare synthetic data

```python
# Reload noisy samples:
samples = np.load('../data/synthetic/samples.npy')
n_samples = 250 # Specify number of samples per class in total.
train_val_split = 0.8 # Choose ratio of train to validation samples.

# Get mask of valid and invalid grid points:
invalid_gridpoints = np.isnan(samples)
valid_gridpoints = (np.isnan(samples) == False)

# Set former NaN values to zero:
samples[invalid_gridpoints] = 0

# Split samples into training and validation sets:
n_train = int(n_samples * train_val_split)
train_input = np.concatenate([samples[:n_train], samples[n_samples:n_samples+n_train]], axis=0)
val_input = np.concatenate([samples[n_train:n_samples], samples[n_samples+n_train:]], axis=0)

# Need to swap samples' axes to feed samples col-wise into ESN model:
train_input_swap = np.swapaxes(train_input, 1, 2)
val_input_swap = np.swapaxes(val_input, 1, 2)

# Keep copy of raw inputs before adding dummy column, and add dimension for number of channels, required for Conv2D:
train_input_raw = np.expand_dims(np.copy(train_input), axis=-1)
val_input_raw = np.expand_dims(np.copy(val_input), axis=-1)

# Add dummy column to swapped and un-swapped inputs containing only ONEs:
train_input = np.concatenate([np.ones((train_input.shape[0],1,100)), train_input], axis=1)
val_input = np.concatenate([np.ones((val_input.shape[0],1,100)), val_input], axis=1)
train_input_swap = np.concatenate([np.ones((train_input_swap.shape[0],1,100)), train_input_swap], axis=1)
val_input_swap = np.concatenate([np.ones((val_input_swap.shape[0],1,100)), val_input_swap], axis=1)    

# Now only keep valid inputs and flatten inputs, then restore first dimension as "samples" 
# and second dimension as "time step" by reshaping. Then split samples into train and validation data:
inputs_flat = np.reshape(samples[valid_gridpoints], (samples.shape[0], 1, np.sum(valid_gridpoints[0].astype(int))))
train_input_flat = np.concatenate([inputs_flat[:n_train], inputs_flat[n_samples:n_samples+n_train]], axis=0)
val_input_flat = np.concatenate([inputs_flat[n_train:n_samples], inputs_flat[n_samples+n_train:]], axis=0)

### Reload random permutation:
rand_perm = np.load('../data/synthetic/rand_perm.npy')

# Apply permutation to flattened inputs:
train_input_flat_perm = train_input_flat[:,:,rand_perm]
val_input_flat_perm = val_input_flat[:,:,rand_perm]

# Split permuted flattened inputs, consisting of only valid grid points, into equal sized chunks:
train_input_chunk = np.reshape(train_input_flat_perm,(400,96,100))
val_input_chunk = np.reshape(val_input_flat_perm,(100,96,100))

# Add dummy column to chunked inputs containing only ONEs:
train_input_chunk = np.concatenate([np.ones((train_input_chunk.shape[0],1,100)), train_input_chunk], axis=1)
val_input_chunk = np.concatenate([np.ones((val_input_chunk.shape[0],1,100)), val_input_chunk], axis=1)

# Create targets: 1 for class1 and -1 for class2
train_target = np.concatenate([np.ones((n_train,1)), -np.ones((n_train,1))])
val_target = np.concatenate([np.ones((n_samples-n_train,1)), -np.ones((n_samples-n_train,1))])               

# Check dimensions:
print("train_input: ", train_input.shape)
print("train_input_raw: ", train_input_raw.shape)
print("train_input_swap: ", train_input_swap.shape)
print("train_input_flat: ", train_input_flat.shape)
print("train_input_chunk: ", train_input_chunk.shape)
print("\nval_input: ", val_input.shape)
print("val_input_raw: ", val_input_raw.shape)
print("val_input_swap: ", val_input_swap.shape)
print("val_input_flat: ", val_input_flat.shape)
print("val_input_chunk: ", val_input_chunk.shape)
print("\ntrain_target: ", train_target.shape)
print("val_target: ", val_target.shape)
```

### LRP with Echo State Networks (ESNs)

Here we use a leaky integrator ESN.

#### Base parameter setup

* n_res = 300
* W_in_lim = 0.1
* leak_rate = 0.005
* spec_radius = 0.8
* sparsity = 0.3

Use these parameters on synthetic input samples as inputs and related targets, together with **tanh** activation.

#### Outline for this part:

- Not only feed samples **column-by-column** into our ESN model, but also **row-wise**. Needed to add a dummy column of ONEs as first column to all samples to absorb residual relevance. 
- Apply new technique input samples split into equal-sized **chunks**. Hereby, only consider valid gridpoints.

**Note:** Feeding samples **column-wise** requires **swapped longitude and latitude**. 

### ESN (column-wise): Mean relevance maps for class 1 and class 2 samples

```python
### Define parameters for ESN model upfront
verbose = True
input_length = train_input_swap.shape[1] # Number of input time steps, directly taken from inputs' shape.
n_layers = 1 # Number of ESN layers in the model.
n_res = 300 # Number of reservoir units.
W_in_lim = 0.1 # Initialize input weights from random uniform distribution in [- W_in_lim, + W_in_lim]
leak_rate = 0.005 # Leak rate used in transition function of reservoir states.
leak_rate_first_step_YN = True # If true, multiply with alpha already in calculating first timestes's res. states.
leaky_integration_YN = True # If True, multiply previous time steps' reservoir states with (1-a).
                            # If False, omit multiplication with (1-a) in reservoir state transition. 
                            # But in any case still multiply new time steps' input (and reservoir recurrence) 
                            # with leakrate a after activation.
activation = 'tanh' # Desired activation function to be used in calculating reservoir state transition.
spec_radius = 0.8 # Spectral radius, becomes largest Eigenvalue of reservoir weight matrix.
sparsity = 0.3 # Sparsity of reservoir weight matrix.
in_features = train_input_swap.shape[2] # Number of input features, directly taken from inputs' shape.
out_features = 1 # Single target value.
```

```python
# Get complete model (output = target prediction) plus short model (output final reservoir states from all layers)
# and all_states (= another shortened model that gives reservoir states for ALL time steps for all inputs).
model, model_short, all_states = setESN(input_length=input_length, in_features=in_features,
                                        out_features=out_features, n_layers=n_layers,
                                        n_res=n_res, W_in_lim=W_in_lim, leak_rate=leak_rate,
                                        leak_rate_first_step_YN=leak_rate_first_step_YN,
                                        leaky_integration_YN = leaky_integration_YN,
                                        activation=activation, spec_radius=spec_radius,
                                        sparsity=sparsity, verbose=verbose)

# Train ESN model's output weights and bias
model = trainESN(model, model_short, train_input_swap, train_target, verbose=verbose)

# Get predictions from trained ESN model and evaluation metrics on model performance:
train_pred, val_pred, train_mae, val_mae, train_mse, val_mse = predESN(model, train_input_swap, val_input_swap,
                                                                       train_target, val_target,
                                                                       verbose=verbose)
```

```python
## Get model predictions on train and validation data:
train_pred_ESN = model.predict(train_input_swap[:,:,:])
val_pred_ESN = model.predict(val_input_swap[:,:,:])

## Get train and validation accuracy: Rel. amount of correctly predicted events.
## NOTE: We use kind of a one-dimensional soft-maxing. Having a positive / negative prediction value relates
## to class 1 and 2, respectively. In other words: Convert prediction to binary prediction.
val_acc_class1_ESN = sum(((val_pred_ESN>0)==(val_target>0)) & (val_target>0)) / sum((val_target>0))
val_acc_class2_ESN = sum(((val_pred_ESN<0)==(val_target<0)) & (val_target<0)) / sum((val_target<0))
train_acc_class1_ESN = sum(((train_pred_ESN>0)==(train_target>0)) & (train_target>0)) / sum((train_target>0))
train_acc_class2_ESN = sum(((train_pred_ESN<0)==(train_target<0)) & (train_target<0)) / sum((train_target<0))

# Print accuracies:
print("\nESN (column-wise), accuracy on synthetic samples")
print("\ntrain_acc_class1", np.round(train_acc_class1_ESN[0],2))
print("train_acc_class2", np.round(train_acc_class2_ESN[0],2))
print("\nval_acc_class1", np.round(val_acc_class1_ESN[0],2))
print("val_acc_class2", np.round(val_acc_class2_ESN[0],2))

# Print loss:
print("\ntrain loss (mse): ", np.round(np.mean((train_target - train_pred_ESN)**2),5))
print("val loss (mse): ", np.round(np.mean((val_target - val_pred_ESN)**2),3))

# Fidelity check: Plot val_pred vs. val_targets
plt.figure(figsize=(16,8))
plt.plot(range(len(val_target[:])),val_target[:],'b',label="true data", alpha=0.3)
plt.plot(range(len(val_pred_ESN[:])),val_pred_ESN[:],'k',  alpha=0.8, label='pred ESN')
plt.title('Fidelity check on VALIDATION data', fontsize=16)
plt.xlabel('timestep', fontsize=14)
plt.ylabel('sst anomaly index', fontsize=14)
plt.legend(fontsize=14)
plt.show()
```

```python
### LRP with ESN (column-wise) on ALL class1 samples (from train inputs)

verbose=False

## Split train inputs into class1 / class2 samples, according to binary target:
samples_class1= train_input_swap[(train_target==1)[:,0]]
samples_class2 = train_input_swap[(train_target==-1)[:,0]]

## Split targets accordingly:
targets_class1 = train_target[(train_target==1)[:,0]]

if verbose:
    print("samples_class1 shape: ", samples_class1.shape)
    print("samples_class2 shape: ", samples_class2.shape)

# Extract number of time steps (timesteps) and variable T to loop over.
# Note: Subtract 1 from absolute number of timesteps to get T, since loops start at 0.
timesteps = samples_class1.shape[1]
T = timesteps - 1

# Extract number of input features:
in_feat = samples_class1.shape[2]

# Rename leak_rate to 'a' for convenience and readability of LRP algorithm:
a = leak_rate

### Extract model weights from trained model:

## Get all weights:
model_weights = np.array(model.get_weights())

## Set input / reservoir / output weights and bias(es):
W_in = model_weights[0]
b_in = model_weights[1]
W_res = model_weights[2]
b_res = model_weights[3]
W_out = model_weights[4]
b_out = model_weights[5]

## Initialize storage for mean relevance over all ElNino samples.
R_all_mean = np.zeros((T+1, in_feat))
    
#Loop over El Nino samples:
for sample_id in range(len(samples_class1)):
    
    # Print status:
    print ("sample: ", sample_id+1, " from ", len(samples_class1))
    
    # Get current sample:
    sample = samples_class1[sample_id:sample_id+1]
    
    # Have trained model as pre-requisite. Check prediction and target on picked sample:
    pred = model.predict(sample)[0,0]
    
    ## Set tensors for inputs U, activation layer ACT and all reservoir states X:

    # Get inputs U for all timesteps from raw input sample by just omitting first dimension (=sample number):
    U = sample[0,:,:]

    # Get all reservoir states for all timesteps X on picked sample, omit first dimension (=sample number):
    X = all_states.predict(sample)[0,:,:]

    # Get X' from all reservoir states X, shifted by one timestep and adding ZEROs as first row:
    X_prime = np.concatenate([np.zeros((1,n_res)),X[:-1,:]], axis=0)

    # Get activation layer ACT from relation: X = (1 - alpha) * X_prime + alpha * ACT
    ACT = (X - (1 - a) * X_prime) / a
    
    ## LRP: Only consider positive pre-activations for propagation rule.

    ## Start with layer T

    # Get relevances R_i and first part of R_j for n_res units x_i in ACT(T) and x_j in X(T-1), respectively,
    # connecting to single OUTPUT unit x_o:
    R_i = (a * (np.multiply(ACT[T,:],W_out[:,0])>0) * np.multiply(ACT[T,:],W_out[:,0]) /
           sum((np.multiply(ACT[T,:],W_out[:,0])>0) * np.multiply(ACT[T,:],W_out[:,0])) * pred)
    R_j_1 = ((1 - a) * (np.multiply(ACT[T,:],W_out[:,0])>0) * np.multiply(ACT[T,:],W_out[:,0])
             / sum((np.multiply(ACT[T,:],W_out[:,0])>0) * np.multiply(ACT[T,:],W_out[:,0])) * pred)

    # Get relevances R_k for in_feat input units x_k of U(T)
    R_k = np.sum((np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in)
                 / (np.sum((np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in), axis=0)
                    + np.sum((np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res), axis=0)) * R_i, axis=1)

    # Get second contribution R_j_2 for relevance R_j for n_res units x_j of X(T-1) connecting to x_i units of ACT(T).
    R_j_2 = np.sum((np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res)
    / (np.sum((np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in), axis=0)
                    + np.sum((np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res), axis=0)) * R_i, axis=1)

    # Combine both contributions for relevance R_j, for n_res units x_j of X(T-1):
    R_j = R_j_1 + R_j_2

    ## Store relevance for input layer U for current time step.
    R_all = np.reshape(R_k,(1,len(R_k)))

    ### Now loop over remaining T-1 layers in decreasing order. Start with T-1, stop at 1, since we need
    ### special treatment there!


    for t in np.flip(np.arange(1,T)):

        # Get relevances R_i for n_res units x_i of ACT(t). Note: Each unit x_i is only connected to a
        # single corresponding unit x_m of X(t), multiplied by alpha. 
        # No additional weights and bias in this connection, hence w=1 and b=0.
        R_i = a * R_j

        # The same holds true for first part of relevances R_j_1 for n_res units x_j of X(t-1) connecting to
        # n_res units x_m of X(t), multiplied by (1-alpha), no weights and bias.
        R_j_1 = (1 - a) * R_j

        # Get relevances R_k for in_feat input units x_k of U(T)
        R_k = np.sum((np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in)
                     / (np.sum((np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in), axis=0)
                        + np.sum((np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res), axis=0)) * R_i, axis=1)

        # Get second contribution R_j_2 for relevance R_j for n_res units x_j of X(T-1) connecting to x_i units of ACT(T).
        R_j_2 = np.sum((np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res)
        / (np.sum((np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in), axis=0)
                        + np.sum((np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res), axis=0)) * R_i, axis=1)

        # Combine both contributions for relevance R_j, for n_res units x_j of X(t-1):
        R_j = R_j_1 + R_j_2

        ## Store relevance for input layer U for current time step.
        R_all = np.concatenate([np.reshape(R_k,(1,len(R_k))), R_all], axis=0)

        
    ## Now take care of time step t=1: This time step is related to the dummy column of ONEs, we added to the
    ## input samples as first column to absorb residual relevance in order to preserve total relevance
    ## without influencing prediction accuracy.

    # Note: There is no previous time step's reservoir states X(t=0).   
    # Hence ACT(1) equals X(1). No multiplication with alpha!
    R_i = R_j

    # And lastly the residual relevance R_i is completely distributed on first time step's inputs U(1):
    R_k = np.sum((np.multiply(np.reshape(U[0,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[0,:],(in_feat,1)),W_in)
                     / (np.sum((np.multiply(np.reshape(U[0,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[0,:],(in_feat,1)),W_in), axis=0)) * R_i, axis=1)

    ## Store relevance for input layer U for very first time step (t=1).
    R_all = np.concatenate([np.reshape(R_k,(1,len(R_k))), R_all], axis=0)
    
    # Normalize relevance map to have sum of total relevance equal to ONE.
    # Simply divide by absolute prediction, which served as total relevance to start with:
    R_all /= abs(pred)
        
    ## Add relevance for current sample to storage for ALL samples:
    R_all_mean += R_all

## Get mean relevance over ALL samples:
R_all_mean = R_all_mean / len(samples_class1)

## Since we worked with swapped sample (lat <-> lon), reverse swapping:
R_all_mean_class1 = np.swapaxes(R_all_mean, 0, 1)

# Set former NaN values to NaN, to highlight land masses:
R_all_mean_class1[:,1:][invalid_gridpoints[0]] = float("nan")
```

```python
### LRP with baseESN (column-wise) on ALL class2 samples (from train inputs)

verbose=False

## Split train inputs into class1 / class2 samples, according to binary target:
samples_class1 = train_input_swap[(train_target==1)[:,0]]
samples_class2 = train_input_swap[(train_target==-1)[:,0]]

## Split targets accordingly:
targets_class2 = train_target[(train_target==-1)[:,0]]

if verbose:
    print("samples_class1 shape: ", samples_class1.shape)
    print("samples_class2 shape: ", samples_class2.shape)

# Extract number of time steps (timesteps) and variable T to loop over.
# Note: Subtract 1 from absolute number of timesteps to get T, since loops start at 0.
timesteps = samples_class2.shape[1]
T = timesteps - 1

# Extract number of input features:
in_feat = samples_class2.shape[2]

# Rename leak_rate to 'a' for convenience and readability of LRP algorithm:
a = leak_rate

### Extract model weights from trained model:

## Get all weights:
model_weights = np.array(model.get_weights())

## Set input / reservoir / output weights and bias(es):
W_in = model_weights[0]
b_in = model_weights[1]
W_res = model_weights[2]
b_res = model_weights[3]
W_out = model_weights[4]
b_out = model_weights[5]

## Initialize storage for mean relevance over all LaNina samples.
R_all_mean = np.zeros((T+1, in_feat))
    
#Loop over El Nino samples:
for sample_id in range(len(samples_class2)):
    
    # Print status:
    print ("sample: ", sample_id+1, " from ", len(samples_class2))
    
    # Get current sample:
    sample = samples_class2[sample_id:sample_id+1]
    
    # Have trained model as pre-requisite. Check prediction and target on picked sample:
    pred = model.predict(sample)[0,0]
    
    ## Set tensors for inputs U, activation layer ACT and all reservoir states X:

    # Get inputs U for all timesteps from raw input sample by just omitting first dimension (=sample number):
    U = sample[0,:,:]

    # Get all reservoir states for all timesteps X on picked sample, omit first dimension (=sample number):
    X = all_states.predict(sample)[0,:,:]

    # Get X' from all reservoir states X, shifted by one timestep and adding ZEROs as first row:
    X_prime = np.concatenate([np.zeros((1,n_res)),X[:-1,:]], axis=0)

    # Get activation layer ACT from relation: X = (1 - alpha) * X_prime + alpha * ACT
    ACT = (X - (1 - a) * X_prime) / a
    
    ## LRP: Only consider negative pre-activations for propagation rule on absolute prediction.
    
    ## Start with layer T

    # Get relevances R_i and first part of R_j for n_res units x_i in ACT(T) and x_j in X(T-1), respectively,
    # connecting to single OUTPUT unit x_o:
    R_i = (a * (np.multiply(ACT[T,:],W_out[:,0])<0) * np.multiply(ACT[T,:],W_out[:,0]) /
           sum((np.multiply(ACT[T,:],W_out[:,0])<0) * np.multiply(ACT[T,:],W_out[:,0])) * abs(pred))
    R_j_1 = ((1 - a) * (np.multiply(ACT[T,:],W_out[:,0])<0) * np.multiply(ACT[T,:],W_out[:,0])
             / sum((np.multiply(ACT[T,:],W_out[:,0])<0) * np.multiply(ACT[T,:],W_out[:,0])) * abs(pred))

    # Get relevances R_k for in_feat input units x_k of U(T)
    R_k = np.sum((np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in)<0) * np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in)
                 / (np.sum((np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in)<0) * np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in), axis=0)
                    + np.sum((np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res)<0) * np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res), axis=0)) * R_i, axis=1)

    # Get second contribution R_j_2 for relevance R_j for n_res units x_j of X(T-1) connecting to x_i units of ACT(T).
    R_j_2 = np.sum((np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res)<0) * np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res)
    / (np.sum((np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in)<0) * np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in), axis=0)
                    + np.sum((np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res)<0) * np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res), axis=0)) * R_i, axis=1)

    # Combine both contributions for relevance R_j, for n_res units x_j of X(T-1):
    R_j = R_j_1 + R_j_2

    ## Store relevance for input layer U for current time step.
    R_all = np.reshape(R_k,(1,len(R_k)))

    ### Now loop over remaining T-1 layers in decreasing order. Start with T-1, stop at 1, since we need
    ### special treatment there!


    for t in np.flip(np.arange(1,T)):

        # Get relevances R_i for n_res units x_i of ACT(t). Note: Each unit x_i is only connected to a
        # single corresponding unit x_m of X(t), multiplied by alpha. 
        # No additional weights and bias in this connection, hence w=1 and b=0.
        R_i = a * R_j

        # The same holds true for first part of relevances R_j_1 for n_res units x_j of X(t-1) connecting to
        # n_res units x_m of X(t), multiplied by (1-alpha), no weights and bias.
        R_j_1 = (1 - a) * R_j

        # Get relevances R_k for in_feat input units x_k of U(T)
        R_k = np.sum((np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in)<0) * np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in)
                     / (np.sum((np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in)<0) * np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in), axis=0)
                        + np.sum((np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res)<0) * np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res), axis=0)) * R_i, axis=1)

        # Get second contribution R_j_2 for relevance R_j for n_res units x_j of X(T-1) connecting to x_i units of ACT(T).
        R_j_2 = np.sum((np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res)<0) * np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res)
        / (np.sum((np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in)<0) * np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in), axis=0)
                        + np.sum((np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res)<0) * np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res), axis=0)) * R_i, axis=1)

        # Combine both contributions for relevance R_j, for n_res units x_j of X(t-1):
        R_j = R_j_1 + R_j_2

        ## Store relevance for input layer U for current timestep.
        R_all = np.concatenate([np.reshape(R_k,(1,len(R_k))), R_all], axis=0)

        
    ## Now take care of time step t=1: This time step is related to the dummy column of ONEs, we added to the
    ## input samples as first column to absorb residual relevance in order to preserve total relevance
    ## without influencing prediction accuracy.

    # Note: There is no previous time step's reservoir states X(t=0).   
    # Hence ACT(1) equals X(1). No multiplication with alpha!
    R_i = R_j

    # And lastly the residual relevance R_i is completely distributed on first time step's inputs U(1):
    R_k = np.sum((np.multiply(np.reshape(U[0,:],(in_feat,1)),W_in)<0) * np.multiply(np.reshape(U[0,:],(in_feat,1)),W_in)
                     / (np.sum((np.multiply(np.reshape(U[0,:],(in_feat,1)),W_in)<0) * np.multiply(np.reshape(U[0,:],(in_feat,1)),W_in), axis=0)) * R_i, axis=1)

    ## Store relevance for input layer U for very first time step (t=1).
    R_all = np.concatenate([np.reshape(R_k,(1,len(R_k))), R_all], axis=0)
    
    # Normalize relevance map to have sum of total relevance equal to ONE.
    # Simply divide by absolute prediction, which served as total relevance to start with:
    R_all /= abs(pred)
        
    ## Add relevance for current sample to storage for ALL samples:
    R_all_mean += R_all

## Get mean relevance over ALL samples:
R_all_mean = R_all_mean / len(samples_class2)

## Since we worked with swapped sample (lat <-> lon), reverse swapping:
R_all_mean_class2 = np.swapaxes(R_all_mean, 0, 1)

# Set former NaN values to NaN, to highlight land masses:
R_all_mean_class2[:,1:][invalid_gridpoints[0]] = float("nan")
```

```python
### LRP with ESN (row-wise) on ALL class 1 samples (from train inputs)

verbose=False

## Split train inputs into El Nino / La Nina samples, according to binary target:
samples_class1 = train_input[(train_target==1)[:,0]]
samples_class2 = train_input[(train_target==-1)[:,0]]

if verbose:
    print("samples_class1 shape: ", samples_class1.shape)
    print("samples_class2 shape: ", samples_class2.shape)

# Extract number of time steps (timesteps) and variable T to loop over.
# Note: Subtract 1 from absolute number of timesteps to get T, since loops start at 0.
timesteps = samples_class1.shape[1]
T = timesteps - 1

# Extract number of input features:
in_feat = samples_class1.shape[2]

# Rename leak_rate to 'a' for convenience and readability of LRP algorithm:
a = leak_rate

### Extract model weights from trained model:

## Get all weights:
model_weights = np.array(model.get_weights())

## Set input / reservoir / output weights and bias(es):
W_in = model_weights[0]
b_in = model_weights[1]
W_res = model_weights[2]
b_res = model_weights[3]
W_out = model_weights[4]
b_out = model_weights[5]

## Initialize storage for mean relevance over all ElNino samples.
R_all_mean = np.zeros((T+1, in_feat))
    
#Loop over El Nino samples:
for sample_id in range(len(samples_class1)):
    
    # Print status:
    print ("sample: ", sample_id+1, " from ", len(samples_class1))
    
    # Get current sample:
    sample = samples_class1[sample_id:sample_id+1]
    
    # Have trained model as pre-requisite. Check prediction and target on picked sample:
    pred = model.predict(sample)[0,0]
    
    ## Set tensors for inputs U, activation layer ACT and all reservoir states X:

    # Get inputs U for all timesteps from raw input sample by just omitting first dimension (=sample number):
    U = sample[0,:,:]

    # Get all reservoir states for all timesteps X on picked sample, omit first dimension (=sample number):
    X = all_states.predict(sample)[0,:,:]

    # Get X' from all reservoir states X, shifted by one timestep and adding ZEROs as first row:
    X_prime = np.concatenate([np.zeros((1,n_res)),X[:-1,:]], axis=0)

    # Get activation layer ACT from relation: X = (1 - alpha) * X_prime + alpha * ACT
    ACT = (X - (1 - a) * X_prime) / a
    
    ## LRP: Only consider positive pre-activations for propagation rule.

    ## Start with layer T

    # Get relevances R_i and first part of R_j for n_res units x_i in ACT(T) and x_j in X(T-1), respectively,
    # connecting to single OUTPUT unit x_o:
    R_i = (a * (np.multiply(ACT[T,:],W_out[:,0])>0) * np.multiply(ACT[T,:],W_out[:,0]) /
           sum((np.multiply(ACT[T,:],W_out[:,0])>0) * np.multiply(ACT[T,:],W_out[:,0])) * pred)
    R_j_1 = ((1 - a) * (np.multiply(ACT[T,:],W_out[:,0])>0) * np.multiply(ACT[T,:],W_out[:,0])
             / sum((np.multiply(ACT[T,:],W_out[:,0])>0) * np.multiply(ACT[T,:],W_out[:,0])) * pred)

    # Get relevances R_k for in_feat input units x_k of U(T)
    R_k = np.sum((np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in)
                 / (np.sum((np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in), axis=0)
                    + np.sum((np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res), axis=0)) * R_i, axis=1)

    # Get second contribution R_j_2 for relevance R_j for n_res units x_j of X(T-1) connecting to x_i units of ACT(T).
    R_j_2 = np.sum((np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res)
    / (np.sum((np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in), axis=0)
                    + np.sum((np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res), axis=0)) * R_i, axis=1)

    # Combine both contributions for relevance R_j, for n_res units x_j of X(T-1):
    R_j = R_j_1 + R_j_2

    ## Store relevance for input layer U for current time step.
    R_all = np.reshape(R_k,(1,len(R_k)))

    ### Now loop over remaining T-1 layers in decreasing order. Start with T-1, stop at 1, since we need
    ### special treatment there!


    for t in np.flip(np.arange(1,T)):

        # Get relevances R_i for n_res units x_i of ACT(t). Note: Each unit x_i is only connected to a
        # single corresponding unit x_m of X(t), multiplied by alpha. 
        # No additional weights and bias in this connection, hence w=1 and b=0.
        R_i = a * R_j

        # The same holds true for first part of relevances R_j_1 for n_res units x_j of X(t-1) connecting to
        # n_res units x_m of X(t), multiplied by (1-alpha), no weights and bias.
        R_j_1 = (1 - a) * R_j

        # Get relevances R_k for in_feat input units x_k of U(T)
        R_k = np.sum((np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in)
                     / (np.sum((np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in), axis=0)
                        + np.sum((np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res), axis=0)) * R_i, axis=1)

        # Get second contribution R_j_2 for relevance R_j for n_res units x_j of X(T-1) connecting to x_i units of ACT(T).
        R_j_2 = np.sum((np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res)
        / (np.sum((np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in), axis=0)
                        + np.sum((np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res), axis=0)) * R_i, axis=1)

        # Combine both contributions for relevance R_j, for n_res units x_j of X(t-1):
        R_j = R_j_1 + R_j_2

        ## Store relevance for input layer U for current timestep.
        R_all = np.concatenate([np.reshape(R_k,(1,len(R_k))), R_all], axis=0)

        
    ## Now take care of time step t=1: This time step is related to the dummy column of ONEs, we added to the
    ## input samples as first column to absorb residual relevance in order to preserve total relevance
    ## without influencing prediction accuracy.

    # Note: There is no previous time step's reservoir states X(t=0).   
    # Hence ACT(1) equals X(1). No multiplication with alpha!
    R_i = R_j

    # And lastly the residual relevance R_i is completely distributed on first time step's inputs U(1):
    R_k = np.sum((np.multiply(np.reshape(U[0,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[0,:],(in_feat,1)),W_in)
                     / (np.sum((np.multiply(np.reshape(U[0,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[0,:],(in_feat,1)),W_in), axis=0)) * R_i, axis=1)

    ## Store relevance for input layer U for very first time step (t=1).
    R_all = np.concatenate([np.reshape(R_k,(1,len(R_k))), R_all], axis=0)
    
    # Normalize relevance map to have sum of total relevance equal to ONE.
    # Simply divide by absolute prediction, which served as total relevance to start with:
    R_all /= abs(pred)
        
    ## Add relevance for current sample to storage for ALL samples:
    R_all_mean += R_all

## Get mean relevance over ALL samples:
R_all_mean_class1 = R_all_mean / len(samples_class1)

# Set former NaN values to NaN, to highlight land masses:
R_all_mean_class1[1:,:][invalid_gridpoints[0]] = float("nan")
```

```python
### LRP with ESN (chunks) on ALL class1 samples (from train inputs)

verbose=False

## Split train inputs into class1 / class2 samples, according to binary target:
samples_class1= train_input_chunk[(train_target==1)[:,0]]
samples_class2 = train_input_chunk[(train_target==-1)[:,0]]

if verbose:
    print("samples_class1 shape: ", samples_class1.shape)
    print("samples_class2 shape: ", samples_class2.shape)

# Extract number of time steps (timesteps) and variable T to loop over.
# Note: Subtract 1 from absolute number of timesteps to get T, since loops start at 0.
timesteps = samples_class1.shape[1]
T = timesteps - 1

# Extract number of input features:
in_feat = samples_class1.shape[2]

# Rename leak_rate to 'a' for convenience and readability of LRP algorithm:
a = leak_rate

### Extract model weights from trained model:

## Get all weights:
model_weights = np.array(model.get_weights())

## Set input / reservoir / output weights and bias(es):
W_in = model_weights[0]
b_in = model_weights[1]
W_res = model_weights[2]
b_res = model_weights[3]
W_out = model_weights[4]
b_out = model_weights[5]

## Initialize storage for mean relevance over all ElNino samples.
R_all_mean = np.zeros((T+1, in_feat))
    
#Loop over El Nino samples:
for sample_id in range(len(samples_class1)):
    
    # Print status:
    print ("sample: ", sample_id+1, " from ", len(samples_class1))
    
    # Get current sample:
    sample = samples_class1[sample_id:sample_id+1]
    
    # Have trained model as pre-requisite. Check prediction and target on picked sample:
    pred = model.predict(sample)[0,0]
    
    ## Set tensors for inputs U, activation layer ACT and all reservoir states X:

    # Get inputs U for all timesteps from raw input sample by just omitting first dimension (=sample number):
    U = sample[0,:,:]

    # Get all reservoir states for all timesteps X on picked sample, omit first dimension (=sample number):
    X = all_states.predict(sample)[0,:,:]

    # Get X' from all reservoir states X, shifted by one timestep and adding ZEROs as first row:
    X_prime = np.concatenate([np.zeros((1,n_res)),X[:-1,:]], axis=0)

    # Get activation layer ACT from relation: X = (1 - alpha) * X_prime + alpha * ACT
    ACT = (X - (1 - a) * X_prime) / a
    
    ## LRP: Only consider positive pre-activations for propagation rule.

    ## Start with layer T

    # Get relevances R_i and first part of R_j for n_res units x_i in ACT(T) and x_j in X(T-1), respectively,
    # connecting to single OUTPUT unit x_o:
    R_i = (a * (np.multiply(ACT[T,:],W_out[:,0])>0) * np.multiply(ACT[T,:],W_out[:,0]) /
           sum((np.multiply(ACT[T,:],W_out[:,0])>0) * np.multiply(ACT[T,:],W_out[:,0])) * pred)
    R_j_1 = ((1 - a) * (np.multiply(ACT[T,:],W_out[:,0])>0) * np.multiply(ACT[T,:],W_out[:,0])
             / sum((np.multiply(ACT[T,:],W_out[:,0])>0) * np.multiply(ACT[T,:],W_out[:,0])) * pred)

    # Get relevances R_k for in_feat input units x_k of U(T)
    R_k = np.sum((np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in)
                 / (np.sum((np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in), axis=0)
                    + np.sum((np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res), axis=0)) * R_i, axis=1)

    # Get second contribution R_j_2 for relevance R_j for n_res units x_j of X(T-1) connecting to x_i units of ACT(T).
    R_j_2 = np.sum((np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res)
    / (np.sum((np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[T,:],(in_feat,1)),W_in), axis=0)
                    + np.sum((np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[T-1,:],(n_res, 1)),W_res), axis=0)) * R_i, axis=1)

    # Combine both contributions for relevance R_j, for n_res units x_j of X(T-1):
    R_j = R_j_1 + R_j_2

    ## Store relevance for input layer U for current time step.
    R_all = np.reshape(R_k,(1,len(R_k)))

    ### Now loop over remaining T-1 layers in decreasing order. Start with T-1, stop at 1, since we need
    ### special treatment there!


    for t in np.flip(np.arange(1,T)):

        # Get relevances R_i for n_res units x_i of ACT(t). Note: Each unit x_i is only connected to a
        # single corresponding unit x_m of X(t), multiplied by alpha. 
        # No additional weights and bias in this connection, hence w=1 and b=0.
        R_i = a * R_j

        # The same holds true for first part of relevances R_j_1 for n_res units x_j of X(t-1) connecting to
        # n_res units x_m of X(t), multiplied by (1-alpha), no weights and bias.
        R_j_1 = (1 - a) * R_j

        # Get relevances R_k for in_feat input units x_k of U(T)
        R_k = np.sum((np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in)
                     / (np.sum((np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in), axis=0)
                        + np.sum((np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res), axis=0)) * R_i, axis=1)

        # Get second contribution R_j_2 for relevance R_j for n_res units x_j of X(T-1) connecting to x_i units of ACT(T).
        R_j_2 = np.sum((np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res)
        / (np.sum((np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[t,:],(in_feat,1)),W_in), axis=0)
                        + np.sum((np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res)>0) * np.multiply(np.reshape(X[t-1,:],(n_res, 1)),W_res), axis=0)) * R_i, axis=1)

        # Combine both contributions for relevance R_j, for n_res units x_j of X(t-1):
        R_j = R_j_1 + R_j_2

        ## Store relevance for input layer U for current timestep.
        R_all = np.concatenate([np.reshape(R_k,(1,len(R_k))), R_all], axis=0)

        
    ## Now take care of time step t=1: This time step is related to the dummy column of ONEs, we added to the
    ## input samples as first column to absorb residual relevance in order to preserve total relevance
    ## without influencing prediction accuracy.

    # Note: There is no previous time step's reservoir states X(t=0).   
    # Hence ACT(1) equals X(1). No multiplication with alpha!
    R_i = R_j

    # And lastly the residual relevance R_i is completely distributed on first time step's inputs U(1):
    R_k = np.sum((np.multiply(np.reshape(U[0,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[0,:],(in_feat,1)),W_in)
                     / (np.sum((np.multiply(np.reshape(U[0,:],(in_feat,1)),W_in)>0) * np.multiply(np.reshape(U[0,:],(in_feat,1)),W_in), axis=0)) * R_i, axis=1)

    ## Store relevance for input layer U for very first time step (t=1).
    R_all = np.concatenate([np.reshape(R_k,(1,len(R_k))), R_all], axis=0)
    
    # Normalize relevance map to have sum of total relevance equal to ONE.
    # Simply divide by absolute prediction, which served as total relevance to start with:
    R_all /= abs(pred)
        
    ## Add relevance for current sample to storage for ALL samples:
    R_all_mean += R_all

## Get mean relevance over ALL samples:
R_all_mean = R_all_mean / len(samples_class1)
```
