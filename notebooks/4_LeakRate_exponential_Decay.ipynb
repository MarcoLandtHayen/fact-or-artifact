### Mathematical approximation for exponentially decaying relevance with ESN models

Assume having a sample consisting of $T=100$ input time steps.

### Tech preamble

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Plot residual relevance for each time step in backward pass:
fig, ax = plt.subplots(figsize=(5, 4))
plt.title('ESN: Residual relevance in backward pass', fontsize=14)
for a in [0.001,0.005,0.01,0.05,0.1,0.2]:
    plt.plot(np.exp(-a*(99-np.arange(0,100,1))), label=str(a)) 
plt.xlabel('Time step t', fontsize=12)
plt.ylabel('Residual relevance R(t)', fontsize=12)
plt.legend(title='leak rate',loc='center left', bbox_to_anchor=(1, 0.5), fontsize=12)
plt.show()
```

```python
# Compute amount of total relevance absorbed by initial input:
print("Absorbed by initial input u(t=1):\n")
for a in [0.001,0.005,0.01,0.05,0.1,0.2]:
    print(a, ": ", np.round(np.exp(-100*a),3))
```

```python
# Plot rate of change for relevance tracing back through time steps in backward pass:
fig, ax = plt.subplots(figsize=(5, 4))
plt.title('Rate of change for residual relevance', fontsize=14)
for a in [0.001,0.005,0.01,0.05,0.1,0.2]:
    plt.plot(a*np.exp(-a*(99-np.arange(0,100,1))), label=str(a)) 
plt.xlabel('Time step t', fontsize=12)
plt.ylabel('Rate of change', fontsize=12)
plt.legend(title='leak rate',loc='center left', bbox_to_anchor=(1, 0.5), fontsize=12)
plt.show()
```

```python
# Plot Residual Relevance and Rate of Change side by side in one plot:
fig, axes = plt.subplots(1,2,figsize=(12, 4))

# Residual Relevance
axes[0].set_title('ESN: Residual relevance in backward pass', fontsize=14)
for a in [0.001,0.005,0.01,0.05,0.1,0.2]:
    axes[0].plot(np.exp(-a*(99-np.arange(0,100,1))), label=str(a)) 
axes[0].set_xlabel('Time step t', fontsize=12)
axes[0].set_ylabel('Residual relevance R(t)', fontsize=12)
#axes[0].legend(title='leak rate',loc='center left', bbox_to_anchor=(1, 0.5), fontsize=12)

# Rate of Change:
axes[1].set_title('Rate of change for residual relevance', fontsize=14)
for a in [0.001,0.005,0.01,0.05,0.1,0.2]:
    axes[1].plot(a*np.exp(-a*(99-np.arange(0,100,1))), label=str(a)) 
axes[1].set_xlabel('Time step t', fontsize=12)
axes[1].set_ylabel('Rate of change', fontsize=12)
axes[1].legend(title='leak rate',loc='center left', bbox_to_anchor=(1, 0.5), fontsize=12)

plt.show()
```
